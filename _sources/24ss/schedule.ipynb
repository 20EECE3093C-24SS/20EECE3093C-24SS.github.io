```python
import pandas as pd
from IPython.core.display import HTML, Markdown

lectures = pd.read_csv('24ss_schedule.tsv', sep='\t')
lectures = lectures.drop(columns=['week'])
lectures['date']= pd.to_datetime(lectures['date'])
lectures = lectures.set_index('date')
lectures = lectures.fillna('')
```

```python
def add_html_row(date, title, topics, prereading, notices, material):
    if "no class" in title.lower():
        html_str = f'''<tr class="align-top">
                          <th scope="row" class="text-left align-top">{date}<br><span class="badge rounded-pill bg-warning text-dark">Warning</span></th>
                          <td class="text-left align-top">{title}</td>
                          <td class="text-left align-top">{topics}</td>
                          <td class="text-left align-top">{prereading}</td>
                          <td class="text-left align-top">{notices}</td>
                          <td class="text-left align-top">{material}</td>
                        </tr>'''
    else:
        html_str = f'''<tr class="align-top">
                          <th scope="row" class="text-left align-top">{date}</th>
                          <td class="text-left align-top">{title}</td>
                          <td class="text-left align-top">{topics}</td>
                          <td class="text-left align-top">{prereading}</td>
                          <td class="text-left align-top">{notices}</td>
                          <td class="text-left align-top">{material}</td>
                        </tr>'''
    return html_str
```

```python
rows = ''
for ind in lectures.index:
    rows+=add_html_row(ind.strftime('%a, %b %d, %Y'),
                       lectures['title'][ind],
                       lectures['topics'][ind],
                       lectures['prereading'][ind],
                       lectures['notices'][ind],
                       lectures['material'][ind])
```

```python
table_html=f'''
    <table class="table table-striped text-left">
  <thead class="thead-dark">
    <tr>
      <th scope="col" class="text-left">Date</th>
      <th scope="col" class="text-left" width="30%">Title</th>
      <th scope="col" class="text-left" width="30%">Topics</th>
      <th scope="col" class="text-left" width="20%">Pre-Reading</th>
      <th scope="col" class="text-left" width="10%">Notices</th>
      <th scope="col" class="text-left" width="10%">Material</th>
    </tr>
  </thead>
  <tbody>
  {rows}
  </tbody>
</table>
'''
```
